```python
#Data Cleaning
df_cleaned = df.drop_duplicates(subset=['municipality', 'elevation_ft'])
#calculatinng min, max and avarages
city_stats = df.groupby('municipality')['elevation_ft'].agg(['min', 'max', 'mean']).reset_index()
city_stats.columns = ['municipality', 'Min_Elevation', 'Max_Elevation', 'Avg_Elevation']
print(city_stats)
```

```python
#Checking the highest and lowest elevated cities in the world with populations > 100000?
filtered_df = df[df['Population'] > 100000]
highest_elevation = filtered_df[filtered_df['elevation_ft'] == filtered_df['elevation_ft'].max()]
lowest_elevation = filtered_df[filtered_df['elevation_ft'] == filtered_df['elevation_ft'].min()]
```

```python
print("Highest Elevated City(s):")
print(highest_elevation[['name', 'elevation_ft']])

print("\nLowest Elevated City(s):")
print(lowest_elevation[['name', 'elevation_ft']])
```
